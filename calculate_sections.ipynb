```python
import json
import os

count = 0
cit_count = 0
cit_and_prev_count = 0
cit_and_after_count = 0
cit_prev_after_count = 0



# Function that gets the documents availble in a folder
def get_docs(documents_folder):
    files = []
    for file_name in os.listdir(documents_folder):
        file_path = os.path.join(documents_folder, file_name)
        if os.path.isfile(file_path):
            files.append(file_path)
    return files

def get_types(data):
  count = 0
  cit_count = 0
  cit_and_prev_count = 0
  cit_and_after_count = 0
  cit_prev_after_count = 0
  for section in data["sentences"]:
    count += 1
    if "-1" in section and "0" in section and "1" in section:
      cit_prev_after_count += 1
    elif "0" in section and "1" in section:
      cit_and_after_count += 1
    elif "-1" in section and "0" in section:
      cit_and_prev_count += 1
    elif "0" in section:
      cit_count += 1
  return count, cit_count, cit_and_prev_count, cit_and_after_count, cit_prev_after_count


docs = get_docs("ground_truth")
sorted_docs = sorted(docs)

for doc in sorted_docs:
  with open(doc, 'r') as file:
    data = json.load(file)
  new_count, new_cit_count, new_cit_and_prev_count, new_cit_and_after_count, new_cit_prev_after_count = get_types(data)
  count += new_count
  cit_count += new_cit_count
  cit_and_prev_count += new_cit_and_prev_count
  cit_and_after_count += new_cit_and_after_count
  cit_prev_after_count += new_cit_prev_after_count

print(f"Ground truth: {count}")
print(f"cit_count: {cit_count}")
print(f"cit_and_prev_count: {cit_and_prev_count}")
print(f"cit_and_after_count: {cit_and_after_count}")
print(f"cit_prev_after_count: {cit_prev_after_count}")


docs = get_docs("with_nli_labels")
sorted_docs = sorted(docs)

count = 0
cit_count = 0
cit_and_prev_count = 0
cit_and_after_count = 0
cit_prev_after_count = 0

for doc in sorted_docs:
  with open(doc, 'r') as file:
    data = json.load(file)
  new_count, new_cit_count, new_cit_and_prev_count, new_cit_and_after_count, new_cit_prev_after_count = get_types(data)
  count += new_count
  cit_count += new_cit_count
  cit_and_prev_count += new_cit_and_prev_count
  cit_and_after_count += new_cit_and_after_count
  cit_prev_after_count += new_cit_prev_after_count

print(f"With NLI: {count}")
print(f"cit_count: {cit_count}")
print(f"cit_and_prev_count: {cit_and_prev_count}")
print(f"cit_and_after_count: {cit_and_after_count}")
print(f"cit_prev_after_count: {cit_prev_after_count}")
```
